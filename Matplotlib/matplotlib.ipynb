### Matplotlib Basics

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
x  = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Plotlib")
```

### Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(y**2,x**2)
plt.subplot(2,2,4)
plt.plot(x**2,y**2)
```

### Object Oriented way

```python
obj = plt.figure()
axis1 = obj.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = obj.add_axes([0.3,0.5,0.3,0.3])
axis2.plot(x,y)
```

### TYPE OF PLOTS

```python
plt.scatter(x,y)
```

```python
plt.hist(y)
```

```python
plt.boxplot(x)
```

### Saving Plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
