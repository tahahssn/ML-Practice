```python
import numpy as np
import seaborn as sns
```

```python
arr = np.array([2,3,4,5,6,7,9,10,11,14,15,17,18,19,21,23,25,26,201])
arr
```

```python
q1 = np.percentile(arr,25)
q1
```

```python
q2 = np.percentile(arr,50)
q2
```

```python
q3 = np.percentile(arr,75)
q3
```

```python
UO = q3 + 1.5*(q3-q1) 
UO
```

```python
LO = q1 - 1.5*(q3-q1) 
LO
```

```python
l = []
for i in arr:
    if i >= LO and i <= UO:
        l.append(i)
arr2 = np.array(l)
arr2
```
