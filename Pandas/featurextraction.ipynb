```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('anime.csv')
df.head()
```

```python
#make a new column for episode count
```

```python
df.loc[1]['Title']
```

```python
def extract_episode(txt):
  check = False
  data = ""
  for i in txt:
    if i == ')':
      check = False
      return data
    if check == True:
      data += i
    if i == '(':
      check = True
```

```python
df['Episodes'] = df["Title"].apply(extract_episode)
```

```python
df['Episodes'] = df['Episodes'].str.replace("eps","")
```

```python
df['Episodes'] = df['Episodes'].astype(int)
```

```python
df
```

```python
#make a new column for time stamp
```

```python
df.loc[0]['Title']
```

```python
def timestamp(txt):
  check = False
  data = ""
  for i in range(len(txt)):
      if txt[i] == ')':
          for j in range(i+1,i + 20):
            data += txt[j]

          return data

```

```python
df['TimeStamp'] = df['Title'].apply(timestamp)
```

```python
#which anime has the highest score
df[df['Score'] == df['Score'].max()]
```

```python
#which anime has the highest episode count
df[df['Episodes'] == df['Episodes'].max()]
```
