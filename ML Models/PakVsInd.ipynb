```python
from sklearn.preprocessing import StandardScaler
cols = ['player_id', 'runs_scored', 'balls_faced', 'fours', 'sixes',
       'wickets_taken', 'overs_bowled', 'economy_rate']
scalar = StandardScaler()

df_cleaned[cols] = scalar.fit_transform(df_cleaned[cols])
```

```python
df_cleaned.head()
```

```python
from scipy.stats import pearsonr

# ----------------------------------
# Pearson Correlation Calculation
# ----------------------------------

# List of features to check against target
selected_features = ['player_id', 'team', 'runs_scored', 'balls_faced', 'fours', 'sixes',
       'wickets_taken', 'overs_bowled', 'economy_rate',
       'match_result', 'man_of_the_match', 'role_Batsman', 'role_Bowler',
       'role_Wicket-Keeper', 'player_name_Fakhar Zaman',
       'player_name_Hardik Pandya', 'player_name_Haris Rauf',
       'player_name_KL Rahul', 'player_name_Rashid Khan',
       'player_name_Ravindra Jadeja', 'player_name_Shadab Khan',
       'player_name_Shaheen Afridi', 'player_name_Virat Kohli']

correlations = {
    feature: pearsonr(df_cleaned[feature], df_cleaned['strike_rate'])[0]
    for feature in selected_features
}
correlation_df = pd.DataFrame(list(correlations.items()), columns=['Feature', 'Pearson Correlation'])
correlation_df.sort_values(by='Pearson Correlation', ascending=False)
```

```python
cat_features = ['team', 'match_result', 'man_of_the_match', 'role_Batsman', 'role_Bowler',
       'role_Wicket-Keeper', 'player_name_Fakhar Zaman',
       'player_name_Hardik Pandya', 'player_name_Haris Rauf',
       'player_name_KL Rahul', 'player_name_Rashid Khan',
       'player_name_Ravindra Jadeja', 'player_name_Shadab Khan',
       'player_name_Shaheen Afridi', 'player_name_Virat Kohli']
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

alpha = 0.05

df_cleaned['strike_rate_bin'] = pd.qcut(df_cleaned['strike_rate'], q = 4, labels=False)
chi2_results = {}

for col in cat_features:
    contingency = pd.crosstab(df_cleaned[col], df_cleaned['strike_rate_bin'])
    chi2_stat, p_val, _, _ = chi2_contingency(contingency)
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'Decision': decision
    }

chi2_df = pd.DataFrame(chi2_results).T
chi2_df = chi2_df.sort_values(by='p_value')
chi2_df
```
